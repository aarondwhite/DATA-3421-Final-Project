```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', 100)

FILEPATH = 'heart_attack_prediction_dataset.csv'
df = pd.read_csv(FILEPATH)
```

```python
col_name_map = {col: col.replace(' ', '_') for col in df.columns}
df = df.rename(columns=col_name_map)
```

```python
df.shape
```

```python
# No duplicate records
df.loc[df.duplicated()]
```

```python
df.plot(kind='scatter', x='Exercise_Hours_Per_Week', y='BMI')
plt.show()
```

```python
(
    df
    .groupby('Country')['Income']
    .mean()
    .reset_index()
    .plot(kind='barh', x='Country', y='Income', ylabel='', legend=None)
)
plt.title('Average Income by Country')
plt.xlabel('Average Income')
plt.show()
```
